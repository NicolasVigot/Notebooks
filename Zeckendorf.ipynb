```python
def pièces(n):
    L=[1,2]
    while L[-1]+L[-2]<n:
        L.append(L[-1]+L[-2])
    return L
```

```python
pièces(100)
```

```python
def décompose(entier):
    L = pièces(entier)
    décomposition = []
    indices = []
    indice = len(L)-1
    while entier>0:
        pièce = L[indice]
        if pièce <= entier:
            décomposition.append(pièce)
            indices.append(indice+1)
            entier -= pièce
            indice -= 1 #La décomposition ne peut pas contenir deux termes consécutifs
        indice -= 1
    return décomposition, indices
```

```python
décompose(100)
```
