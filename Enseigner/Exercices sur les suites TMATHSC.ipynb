# Manipuler les suites avec Python

## Plusieurs manières de calculer les termes d'une suite

### Définition explicite
$$u_n=n^2+3$$

```python
def u(n):
    return n**2+3
```

```python
[u(n) for n in range(0,11)]
```

```python
for i in range(11):
    print(f"u({i})={u(i)}")
```

### Définition par récurrence
$$u_0=1$$
$$u_{n+1}=1.1*u_n-2$$

```python
# Première méthode
def u(n):
    u=1
    for k in range(n):
        u = 2*u-2
    return u
```

```python
[u(n) for n in range(11)]
```

```python
# Deuxième méthode
def U(n):
    u=[1]
    for k in range(n):
        u.append(2*u[-1]-2)
    return u
```

```python
U(10)
```

```python
# Troisième méthode (récursive)
def u(n):
    if n==0:
        return 1
    else:
        return 2*u(n-1)-2
```

```python
[u(n) for n in range(11)]
```

## Algorithmes de seuils
On conjecture que la suite $u$ diverge vers $-\infty$. Quelle la plus petite valeur de $n$ telle que $u_n < -1.000.000$ ? 

```python
def seuil(s):
    n=0
    while u(n) >= s:
        n += 1
    return n
```

```python
n=seuil(-1000000)
n, u(n), u(n-1)
```
