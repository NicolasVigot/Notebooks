## Algorithme d'Euclide

```python
# Version rﾃｩcursive (la plus jolie)
def pgcd(a,b):
    if b==0:
        return a
    else:
        q=a//b
        r=a%b
        print(f"{a}={b}筧ｯ{q}+{r}")
        return pgcd(b,r)
```

```python
pgcd(17136,12600)
```

```python
# Version itﾃｩrative
def PGCD(a,b):
    while b!=0:
        q=a//b
        r=a%b
        print(f"{a}={b}筧ｯ{q}+{r}")
        a,b=b,r
    return a
```

```python
PGCD(17136,12600)
```

## Nombres premiers entre eux

```python
def premiers(a,b):
    return pgcd(a,b)==1
```

```python
premiers(13,5)
```

## Coefficients de Bezout (mﾃｩthode feignante)

```python
def bezout(a,b):
    assert (a!=0) and (b!=0), f"l'un des paramﾃｨtres est nul."
    assert pgcd(a,b)==1, f"{a} et {b} ne sont pas premiers entre eux."
    
    u=0
    while (1-a*u) % b != 0: 
        u+=1
    v=(1-a*u)//b
    return (u,v)
```

```python
bezout(15,4)
```
