# Random Sierpinsky

Le principe de cet algorithme est le suivant : 
- On part d'un triangle $ABC$ équilatéral.
- On place un point $M$ intérieur (en fait, pas nécessairement, mais ça donne un dessin plus régulier et ça borne la figure).
- On choisi au hasard l'un des trois points $A$, $B$ ou $C$. On place le point $M_{n+1}$, au milieu du segment formé par le point choisi et le point $M_n$. 
- On itère le procédé un grand nombre de fois. 

```python
import matplotlib.pyplot as plt
import random
from math import sqrt
```

```python
A,B,C=(0,0),(1,0),(1/2,sqrt(3)/2)
T=[A,B,C]
```

```python
def processus(n):
    '''
    On commence par placer un point "à peu près" à l'intérieur du triangle T 
    '''
    X,Y=[0.5+random.random()/3],[0.5+random.random()/3]
    for _ in range(n):
        P = random.choice(T)
        X.append((X[-1]+P[0])/2)
        Y.append((Y[-1]+P[1])/2)
    return X,Y
```

```python
n = 100000
```

```python
X,Y = processus(n)
plt.clf()
plt.figure(figsize=(10, 10))
plt.plot(X, Y, "b.", markersize=1)
plt.plot([A[0], B[0], C[0], A[0]], [A[1], B[1], C[1], A[1]],"r-", linewidth=5)
plt.axis("equal")
plt.grid(False)
plt.show()
```
