## Algorithme d'Euclide

```python
def pgcd(a,b):
    if b==0:
        return a
    else:
        print(f"{a}={b}⨯{a//b}+{a%b}")
        return pgcd(b,a%b)
```

```python
pgcd(17136,12600)
```

## Nombres premiers entre eux

```python
def premiers(a,b):
    return pgcd(a,b)==1
```

```python
premiers(13,5)
```

## Coefficients de Bezout (méthode feignante)

```python
def bezout(a,b):
    assert (a!=0) and (b!=0), f"l'un des paramètres est nul."
    assert pgcd(a,b)==1, f"{a} et {b} ne sont pas premiers entre eux."
    
    u=0
    while (1-a*u) % b != 0: 
        u+=1
    v=(1-a*u)//b
    return (u,v)
```

```python
bezout(15,4)
```
